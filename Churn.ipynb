```python
import pandas as pd
import numpy as np
mini = pd.read_csv('C:/Users/Rakhman/Anaconda3/dataBLE/mini.csv')
(a,b) = mini.shape
v = []
for i in range(a):
    v.append(0)
mini['month']=v
mini.head()
```

```python
def f(m):
    if m[0:4]=='2017':
        return 0
    x = int(m[5:7])
    if x>6:
        return 0
    else:
        return x
mini['month'] = mini['occurence'].apply(f)
mini
```

```python
mini = mini[mini['month']!=0]
```

```python
mini
```

```python
first5months = mini[mini['month']<6]
month6 = mini[mini['month']==6]
```

```python
unique5months = first5months.groupby(by='customer_id').median()
```

```python
unique6month = month6.groupby(by='customer_id').median()
```

```python
unique6month.head()
```

```python
unique5months.head()
```

```python
def fun(x):
    if x in unique5months.index and x in unique6month.index:
        return 0
    if x in unique5months.index and not x in unique6month.index:
        return 1
    return 0.5
mini['churn'] = mini['customer_id'].apply(fun)
```

```python
mini.head()
```

```python
mini = mini[mini['churn']!=0.5]
mini.head()
mini.shape
```

```python
kz = pd.DataFrame(mini.groupby(by='customer_id')['cost'].mean())
```

```python
kz['number of transactions'] = mini.groupby('customer_id')['cost'].count()
kz['total'] = mini.groupby('customer_id')['cost'].sum()
```

```python
kz['churn'] = mini.groupby('customer_id')['churn'].median()
kz
```

```python
from sklearn.model_selection import train_test_split, cross_val_score
```

```python
y = kz['churn']
X = kz.drop('churn', axis=1, inplace=True)
```

```python
y.head()
```

```python
X = kz
```

```python
X.head()
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = 0.3, random_state = 7)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeClassifier
```

```python
first_tree = DecisionTreeClassifier(random_state = 3)
tree_params = {'max_depth':np.arange(1,11)}
```

```python
tree1 = GridSearchCV(first_tree, tree_params, cv=5, n_jobs=-1)
```

```python
tree1.fit(X_train, y_train)
```

```python
tree1.best_score_, tree1.best_params_
```

```python
tree_pred = tree1.predict(X_valid)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_valid, tree_pred)
```
