```python
import pandas as pd
import datetime as dt
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r'C:\Users\muhas\Desktop\data\mini.csv')
```

```python
data.shape
```

```python
print(data['occurence'].min(), data['occurence'].max())
```

```python
data['occurence'] = data['occurence'].apply(pd.to_datetime)
```

```python
data.head()
```

```python
sd = dt.datetime(2019,6,28)
data['hist']=sd - data['occurence']
data['hist'].astype('timedelta64[D]')
data['hist']=data['hist'] / np.timedelta64(1, 'D')
data.head()
```

```python
rfmTable_customer = data.groupby(['item_id', 'customer_id']).agg({'hist': lambda x:x.min(), # Recency
                                        'customer_id': lambda x: len(x),               # Frequency
                                        'cost': lambda x: x.sum()})          # Monetary Value

rfmTable_customer.rename(columns={'hist': 'recency', 
                         'customer_id': 'frequency', 
                         'cost': 'monetary_value'}, inplace=True)
```

```python
rfmTable_item = data.groupby('item_id').agg({'hist': lambda x:x.min(), 
                                        'item_id': lambda x: len(x),               
                                        'cost': lambda x: x.sum()})          

rfmTable_item.rename(columns={'hist': 'recency', 
                         'item_id': 'frequency', 
                         'cost': 'monetary_value'}, inplace=True)
```
